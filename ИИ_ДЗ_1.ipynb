```python
import pandas as pd
import numpy as np

# Реальные средние температуры (примерные данные) для городов по сезонам
seasonal_temperatures = {
    "New York": {"winter": 0, "spring": 10, "summer": 25, "autumn": 15},
    "London": {"winter": 5, "spring": 11, "summer": 18, "autumn": 12},
    "Paris": {"winter": 4, "spring": 12, "summer": 20, "autumn": 13},
    "Tokyo": {"winter": 6, "spring": 15, "summer": 27, "autumn": 18},
    "Moscow": {"winter": -10, "spring": 5, "summer": 18, "autumn": 8},
    "Sydney": {"winter": 12, "spring": 18, "summer": 25, "autumn": 20},
    "Berlin": {"winter": 0, "spring": 10, "summer": 20, "autumn": 11},
    "Beijing": {"winter": -2, "spring": 13, "summer": 27, "autumn": 16},
    "Rio de Janeiro": {"winter": 20, "spring": 25, "summer": 30, "autumn": 25},
    "Dubai": {"winter": 20, "spring": 30, "summer": 40, "autumn": 30},
    "Los Angeles": {"winter": 15, "spring": 18, "summer": 25, "autumn": 20},
    "Singapore": {"winter": 27, "spring": 28, "summer": 28, "autumn": 27},
    "Mumbai": {"winter": 25, "spring": 30, "summer": 35, "autumn": 30},
    "Cairo": {"winter": 15, "spring": 25, "summer": 35, "autumn": 25},
    "Mexico City": {"winter": 12, "spring": 18, "summer": 20, "autumn": 15},
}

# Сопоставление месяцев с сезонами
month_to_season = {12: "winter", 1: "winter", 2: "winter",
                   3: "spring", 4: "spring", 5: "spring",
                   6: "summer", 7: "summer", 8: "summer",
                   9: "autumn", 10: "autumn", 11: "autumn"}

# Генерация данных о температуре
def generate_realistic_temperature_data(cities, num_years=10):
    dates = pd.date_range(start="2010-01-01", periods=365 * num_years, freq="D")
    data = []

    for city in cities:
        for date in dates:
            season = month_to_season[date.month]
            mean_temp = seasonal_temperatures[city][season]
            # Добавляем случайное отклонение
            temperature = np.random.normal(loc=mean_temp, scale=5)
            data.append({"city": city, "timestamp": date, "temperature": temperature})

    df = pd.DataFrame(data)
    df['season'] = df['timestamp'].dt.month.map(lambda x: month_to_season[x])
    return df

# Генерация данных
data = generate_realistic_temperature_data(list(seasonal_temperatures.keys()))
data.to_csv('temperature_data.csv', index=False)
```

```python
data.describe()
```

```python
data['year'] = data['timestamp'].dt.year
data['month'] = data['timestamp'].dt.month
```

```python
%%time
data['rolling_mean_30d'] = data.groupby('city')['temperature'].transform(
    lambda x: x.rolling(window=30, center=True).mean()
)
data['rolling_mean_30d'] = data['rolling_mean_30d'].fillna(method='bfill')
data['rolling_mean_30d'] = data['rolling_mean_30d'].fillna(method='ffill')

#Средняя температура и стандартное отклонение для каждого сезона в каждом городе
season_stats = data.groupby(['city', 'season'])['temperature'].agg(['mean', 'std']).reset_index()
season_stats = season_stats.rename(columns={'mean': 'season_mean', 'std': 'season_std'})
data = data.merge(season_stats, on=['city', 'season'], how='left')

#Поиск аномалий
data['is_anomaly'] = abs(data['temperature'] - data['season_mean']) > 2 * data['season_std']
```

```python
import matplotlib.pyplot as plt
# Средняя температура по годам для каждого города
long_term_trends = data.groupby(['city', 'year'])['temperature'].mean().reset_index()

# Визуализация долгосрочных трендов
plt.figure(figsize=(14, 8))
for city in long_term_trends['city'].unique():
    city_data = long_term_trends[long_term_trends['city'] == city]
    plt.plot(city_data['year'], city_data['temperature'], label=city)

plt.title('Долгосрочные тренды изменения температуры')
plt.xlabel('Год')
plt.ylabel('Средняя температура (°C)')
plt.legend()
plt.grid()
plt.show()
```

```python
# Средняя температура по месяцам
seasonal_trends = data.groupby(['city', 'month'])['temperature'].mean().reset_index()

# Визуализация сезонных трендов
plt.figure(figsize=(14, 8))
for city in seasonal_trends['city'].unique():
    city_data = seasonal_trends[seasonal_trends['city'] == city]
    plt.plot(city_data['month'], city_data['temperature'], label=city)

plt.title('Сезонные изменения температуры')
plt.xlabel('Месяц')
plt.ylabel('Средняя температура (°C)')
plt.legend()
plt.grid()
plt.show()
```

```python
# Доля аномалий по городам
anomaly_ratio = data.groupby('city')['is_anomaly'].mean().sort_values()

# Визуализация
anomaly_ratio.plot(kind='bar', figsize=(14, 8), title='Доля аномальных дней по городам')
plt.ylabel('Доля аномалий')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Линейная регрессия для долгосрочных трендов
trend_results = {}
for city in long_term_trends['city'].unique():
    city_data = long_term_trends[long_term_trends['city'] == city]
    X = city_data['year'].values.reshape(-1, 1)
    y = city_data['temperature'].values
    model = LinearRegression().fit(X, y)
    trend_results[city] = model.coef_[0]

# Визуализация трендов
trend_df = pd.DataFrame(list(trend_results.items()), columns=['city', 'trend'])
trend_df = trend_df.sort_values(by='trend', ascending=False)
trend_df.plot(kind='bar', x='city', y='trend', figsize=(14, 8), title='Температурные тренды (°C/год)')
plt.ylabel('Температурный тренд (°C/год)')
plt.show()
```

```python
import multiprocessing as mp

# Разделяем данные по городам
cities = data['city'].unique()

def analyze_city(city):
    city_data = data[data['city'] == city].copy()

    # Скользящее среднее
    city_data['rolling_mean_30d'] = city_data['temperature'].rolling(window=30, center=True).mean()

    # Средняя температура и стандартное отклонение для каждого сезона
    stats = city_data.groupby('season')['temperature'].agg(['mean', 'std']).reset_index()
    stats = stats.rename(columns={'mean': 'season_mean', 'std': 'season_std'})
    city_data = city_data.merge(stats, on='season', how='left')

    # Аномалии
    city_data['is_anomaly'] = abs(city_data['temperature'] - city_data['season_mean']) > 2 * city_data['season_std']
    return city_data
```

```python
%%time
# Параллельная обработка
with mp.Pool(processes=mp.cpu_count()) as pool:
    results = pool.map(analyze_city, cities)

data_parallel = pd.concat(results)
```

Скорость анализа с распараллеливанием было бы выше если было бы намного больше данных, а так получилось дороже создавать потоки.

Скользящее среднее позволило сгладить краткосрочные колебания.

Долгосрочные тренды показывают изменение температуры в разных городах.

Сезонные тренды выявляют характерные колебания температуры по месяцам.

Анализ трендов с использованием регрессии дает оценку изменений температуры для каждого города.

```python
from sklearn.linear_model import LinearRegression

# Расчет долгосрочных трендов с учетом всех данных
all_trends = {}
for city in data['city'].unique():
    city_data = data[data['city'] == city].groupby('year')['temperature'].mean().reset_index()
    X = city_data['year'].values.reshape(-1, 1)
    y = city_data['temperature'].values
    model = LinearRegression().fit(X, y)
    all_trends[city] = model.coef_[0]

# Долгосрочные тренды для данных без аномалий
no_anomaly_trends = {}
for city in data['city'].unique():
    city_data = data[(data['city'] == city) & (~data['is_anomaly'])].groupby('year')['temperature'].mean().reset_index()
    if city_data.shape[0] > 1:  # Проверка, есть ли данные для анализа
        X = city_data['year'].values.reshape(-1, 1)
        y = city_data['temperature'].values
        model = LinearRegression().fit(X, y)
        no_anomaly_trends[city] = model.coef_[0]
    else:
        no_anomaly_trends[city] = None  # Недостаточно данных

# Частота аномалий
anomaly_frequency = data.groupby('city')['is_anomaly'].mean()

# Сравнение трендов
trend_comparison = pd.DataFrame({
    'city': list(all_trends.keys()),
    'trend_with_anomalies': list(all_trends.values()),
    'trend_without_anomalies': [no_anomaly_trends[city] for city in all_trends.keys()],
    'anomaly_frequency': anomaly_frequency.values
})
trend_comparison['trend_difference'] = trend_comparison['trend_with_anomalies'] - trend_comparison['trend_without_anomalies']
```

```python
trend_comparison.dropna().set_index('city')[['trend_with_anomalies', 'trend_without_anomalies']].plot(
    kind='bar', figsize=(14, 8), title='Тренды изменения температуры (с аномалиями и без них)'
)
plt.ylabel('Тренд (°C/год)')
plt.xlabel('Город')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(trend_comparison['anomaly_frequency'], trend_comparison['trend_difference'], c='blue', alpha=0.6)
plt.title('Связь между частотой аномалий и изменением трендов')
plt.xlabel('Частота аномалий')
plt.ylabel('Изменение тренда (°C/год)')
plt.grid()
plt.show()
```
